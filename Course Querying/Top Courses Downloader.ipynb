```python
from googlesearch import search
```

```python
import time
```

```python
df = pd.read_csv('coursedetails.csv')
```

```python
#pip install google-api-python-client
```

```python
df
```

```python
CourseNames = df['CourseName']
```

```python
CourseNames = CourseNames.to_frame()
```

```python
CourseNames = CourseNames.drop_duplicates()
```

```python
UselessCourses = []
for course in CourseNames["CourseName"]:
    if ('Project' in course or 'Laboratory' in course or 'lab' in course or 'Viva' in course or 'Lab' in course or 'Preparatory' in course or 'Seminar' in course):
        UselessCourses.append(course)
```

```python
UselessCourses
```

```python
CourseNames = CourseNames[~CourseNames['CourseName'].isin(UselessCourses)]
```

```python
csv_file = open('CourseSearcher.csv','w',encoding = 'UTF-8')
```

```python
csv_writer = csv.writer(csv_file)
```

```python
csv_writer.writerow(['Title',"Content"])
```

```python
suffix = ' course content'
```

```python
#need to get rid of the excess /n's present in the text of each page. Need code that replaces \n's that occur back to back as a single \n.
```

```python
test = CourseNames[3000:3080]
```

```python
test
```

```python
search_results_all = []
counter = 0
```

```python
test = test[0:30]
```

```python
len(test)
```

```python
for query in test['CourseName']:   
    
    search_result = []
    for i in search(query+suffix,        # The query you want to run
                    tld = 'com',  # The top level domain
                    lang = 'en',  # The language
                    num = 10,     # Number of results per page
                    start = 0,    # First result to retrieve
                    stop = 10,  # Last result to retrieve
                    pause = 2.0,  # Lapse between HTTP requests
                   ):
        search_result.append(i)

    to_be_deleted = []
    for link in search_result:
        if ('.pdf' in link or 'youtube' in link or ".doc" in link or 'download' in link or 'htm?' in link) :
            to_be_deleted.append(link)

    new_results = [entry for entry in search_result if entry not in to_be_deleted]
    search_results_all.append(new_results)
    time.sleep(10)
```
